```python
import pandas as pd
json_info= pd.read_csv("C:/Users/ANIS Saadaoui/Downloads/json_info_ar.txt",sep='delimiter',names=['json_info'])
# json_info1= pd.read_csv("C:/Users/ANIS Saadaoui/json_fr_neg.txt",sep='delimiter',names=['json_info'])
```

```python
json_info
```

```python
json_info.drop(json_info.index[(json_info['json_info']=='none')],axis=0,inplace=True)
```

```python
#extract name from json text
def split_name(text):
    if 'Name' in text:
        a=text.split("'Name':")[1]
        if ',' in a:
            return a.split(",")[0]
        else:
            return a.split('}')[0]
    else:
        return('none')
```

```python
#extract id users from json text
def split_id(text):
    if 'id' in text:
        a=text.split("'id':")[-1]
        if ',' in a:
            return a.split(",")[0]
    else:
        return('none')
```

```python
text= "'id': '100025385132940', 'Name': 'Jaouher Badri','Lieux de résidence':['hammamet']}'"
text.split("'id':")[0]
```

```python
#extract Places Lived of users from json text
def split_ville(text):
    if 'Lieux' in text:
        a=text.split("'Lieux de résidence':")[1]
        if ',' in a:
            return a.split(",")[0]
        else:
            return a.split('}')[0]
    elif 'Places Lived' in text:
        b=text.split("'Places Lived':")[1]
        if ',' in b:
            return b.split(",")[0]
        else:
            return b.split('}')[0]
    else:
        return('none')
```

```python
#exemple
text="'id':'02534684','Name':'anis','Lieux de résidence':'sfax'}"
print(split_ville(text),split_name(text),split_id(text))
```

```python
json_info
```

```python
#apply split_name,split_id,split_ville in all json file
json_info['split_name']=json_info['json_info'].apply(lambda split_info:split_name(split_info))
json_info['split_id']=json_info['json_info'].apply(lambda split_info:split_id(split_info))
json_info['split_ville']=json_info['json_info'].apply(lambda split_info:split_ville(split_info))
```

```python
def split_ville_actuelle(text):
    if "Ville" in text:
        return text.split("Ville")[0]
    elif "Current" in text:
        return text.split("Current")[0]
    else:    
        return text
```

```python
#exemple
text="'Sidi Bou Saïd\nVille actuelle\nKasserine\n'"
split_ville_actuelle(text)
```

```python
def add_n(text):
    if not '\n' in text:
        return text + '\n'
    else:
        return text
json_info['split_ville']=json_info['split_ville'].apply(lambda text:add_n(text))
```

```python
json_info['split_ville']
```

```python
import re
json_info['split_ville_actuel']=json_info['split_ville'].apply(lambda text:split_ville_actuelle(text))
json_info.drop(json_info.index[(json_info['split_ville_actuel']=='none')],axis=0,inplace=True)
json_info['split_ville_actuel'] = json_info['split_ville_actuel'].replace(r'\n',' ', regex=True) 
json_info['split_ville_actuel'] = json_info['split_ville_actuel'].replace("'",'', regex=True) 
json_info['split_ville_actuel'] = json_info['split_ville_actuel'].replace("]",'', regex=True)
json_info['split_ville_actuel'] = json_info['split_ville_actuel'].replace(re.escape("["),'', regex=True)
```

```python
json_info.drop(json_info.index[(json_info['split_ville_actuel']=='none')],axis=0,inplace=True)
```

```python
json_info['split_ville_actuel']
```

```python
json_info['split_ville_actuel']=json_info['split_ville_actuel'].apply(lambda text:text.lower())
```

```python
json_info['split_ville_actuel'].value_counts()
```
